```python
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
m=movies.merge(credits,on='title')
```

```python
m.shape
```

```python
m.info()
```

```python
#things to drop budget,homepage,ol,ot,popul,prod_comp,countries,rd,revenue,runtime
m=m[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
m.head(1)
```

```python
m.isnull().sum()
```

```python
m.dropna(inplace=True)
```

```python
m.duplicated().sum()
```

```python
m.iloc[0].genres
```

```python
import ast
```

```python
def justk(obj):
  ls=[]
  for i in ast.literal_eval(obj):
    ls.append(i['name'])
  return ls
```

```python
m['genres']=m['genres'].apply(justk)
```

```python
m.head()
```

```python
m.iloc[0].keywords
```

```python
m['keywords']=m['keywords'].apply(justk)
```

```python
m.head(1)
```

```python
def justkcast(obj):
  ls=[]
  c=0
  for i in ast.literal_eval(obj):
    if c!=3:
      c+=1
      ls.append(i['name'])
    else:
      break
  return ls
```

```python
m['cast']=m['cast'].apply(justkcast)
```

```python
m.head(1)
```

```python
def dirr(obj):
  ls=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      ls.append(i['name'])
      break
  return ls
```

```python
m['crew']=m['crew'].apply(dirr)
```

```python
m.head(1)
```

```python
m['overview']=m['overview'].apply(lambda x:x.split())
```

```python
m.tail(1)
```

```python
m['genres']=m['genres'].apply(lambda x:[i.replace(' ','') for i in x])
m['cast']=m['cast'].apply(lambda x:[i.replace(' ','') for i in x])
m['crew']=m['crew'].apply(lambda x:[i.replace(' ','') for i in x])
m['keywords']=m['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
m.head()
```

```python
m['tags']=m['overview']+m['genres']+m['keywords']+m['cast']+m['crew']
m.head()
```

```python
df=m[['movie_id','title','tags']]
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head(1)
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vec=cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
df.info
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
fd=cosine_similarity(vec)
```

```python
fd.shape
```

```python
def reco(movie):
  movie_index=df[df['title']==movie].index[0]
  dis=fd[movie_index]
  movies_list=sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[1:10]
  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
reco("Batman Begins")
```
